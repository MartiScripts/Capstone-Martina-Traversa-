```python
import geopy
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

# Inizializza il geocoder con un timeout più lungo
geolocator = Nominatim(user_agent="myGeocoder", timeout=10)  # Aumenta il timeout a 10 secondi

# Funzione per ottenere le coordinate con gestione degli errori
def get_coordinates(city_name):
    try:
        location = geolocator.geocode(city_name + ", Italy")
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        print(f"Timeout durante la ricerca delle coordinate per {city_name}")
        return None, None

# Prova per la città di Torino
lat, lon = get_coordinates("Torino")
if lat and lon:
    print(f"Coordinate di Torino: {lat}, {lon}")
else:
    print("Coordinate non trovate per Torino.")
```

```python
import pandas as pd
import folium

# Carica i dataset
dati_mobilita_urbana = pd.read_csv(r"C:\Users\marti\OneDrive\Desktop\dati_mobilita-urbana.csv", encoding="ISO-8859-1")
correlazioni_mobilita = pd.read_csv(r"C:\Users\marti\OneDrive\Desktop\correlazioni_mobilita_con_coordinate.csv", encoding="ISO-8859-1")

# Unisci i due dataset sulla base della colonna 'Citta'
dati_completi = pd.merge(dati_mobilita_urbana, correlazioni_mobilita, on="Citta")

# Rimuovi le righe con coordinate mancanti
dati_completi = dati_completi.dropna(subset=['Latitudine', 'Longitudine'])

# Crea la mappa centrata sull'Italia (ad esempio, coordinate centrali)
mappa = folium.Map(location=[42.5, 12.5], zoom_start=6)

# Aggiungi i marker sulla mappa
for index, row in dati_completi.iterrows():
    latitudine = row['Latitudine']
    longitudine = row['Longitudine']
    uso_mezzi_pubblici = row['Uso dei mezzi pubblici (%)']

    # Verifica che le coordinate non siano NaN (ulteriore controllo di sicurezza)
    if pd.notna(latitudine) and pd.notna(longitudine):
        # Usa l'uso dei mezzi pubblici per determinare il colore del marker
        color = 'green' if uso_mezzi_pubblici >= 50 else 'red'

        # Aggiungi il marker sulla mappa
        folium.CircleMarker(
            location=[latitudine, longitudine],
            radius=10,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.6,
            popup=f"Città: {row['Citta']}\nUso mezzi pubblici: {uso_mezzi_pubblici}%",
        ).add_to(mappa)

# Salva la mappa come file HTML
mappa.save("mappa_uso_trasporto_pubblico.html")

# Mostra la mappa nel caso stai usando Jupyter
mappa
```

```python
import pandas as pd

# Carica i file CSV
dati_mobilita_urbana = pd.read_csv(r"C:\Users\marti\OneDrive\Desktop\dati_mobilita-urbana.csv", encoding='ISO-8859-1')
dati_trasporto = pd.read_csv(r"C:\Users\marti\OneDrive\Desktop\dati_trasporto_pubblico.csv", encoding='ISO-8859-1')
benefici_ambientali = pd.read_csv(r"C:\Users\marti\OneDrive\Desktop\benefici_ambientali.csv", encoding='ISO-8859-1')

# Verifica le colonne di ciascun dataset
print(dati_mobilita_urbana.columns)
print(dati_trasporto.columns)
print(benefici_ambientali.columns)

# Unisci i dati tra 'benefici_ambientali' e 'dati_mobilita_urbana' sui nomi delle città
dati_completi = pd.merge(benefici_ambientali, dati_mobilita_urbana, on='Citta')

# Unisci anche 'dati_trasporto' per ottenere i km di piste ciclabili
dati_completi = pd.merge(dati_completi, dati_trasporto[['Citta', 'Piste ciclabili (km)']], on='Citta')

# Crea un punteggio di sostenibilità che include CO2, km di piste ciclabili e uso dei mezzi pubblici
# Consideriamo i km di piste ciclabili come una metrica direttamente (senza calcolare percentuali)
dati_completi['Punteggio Sostenibilità'] = (
    dati_completi['Riduzione CO2 con trasporto pubblico (%) (annuo)'] +
    dati_completi['Riduzione CO2 con biciclette (%) (annuo)'] +
    (dati_completi['Piste ciclabili (km)'] / 10) +  # Scala i km di piste ciclabili per il punteggio
    (dati_completi['Uso dei mezzi pubblici (%)'] / 10)  # Usa la percentuale di mezzi pubblici
)

# Ordina le città per punteggio sostenibilità
citta_virtuose = dati_completi[['Citta', 'Punteggio Sostenibilità']].sort_values(by='Punteggio Sostenibilità', ascending=False)

# Mostra le città con il punteggio più alto
print(citta_virtuose)
```

```python
# Filtro per Milano e Roma
citta_comparate = dati_completi[dati_completi['Citta'].isin(['Milano', 'Roma'])]

# Mostra i dati comparati
print(citta_comparate[['Citta', 'Riduzione CO2 con trasporto pubblico (%) (annuo)', 
                       'Riduzione CO2 con biciclette (%) (annuo)', 
                       'Piste ciclabili (km)', 'Uso dei mezzi pubblici (%)']])
```

```python
# Crea un grafico a dispersione del punteggio di sostenibilità rispetto alla popolazione
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dati_completi, x='Popolazione (2023)', y='Punteggio Sostenibilità')

# Aggiungi il titolo e le etichette
plt.title('Punteggio di Sostenibilità rispetto alla Popolazione delle Città', fontsize=16)
plt.xlabel('Popolazione (2023)', fontsize=12)
plt.ylabel('Punteggio di Sostenibilità', fontsize=12)

# Mostra il grafico
plt.tight_layout()
plt.show()
```
